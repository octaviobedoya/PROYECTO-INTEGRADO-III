```python
#Fecha con mayor cancelaciones
ordenes_canceladas = minorista_limpieza_df[minorista_limpieza_df['InvoiceNo'].astype(str).str.contains('C')]

# Contar las cancelaciones por fecha (InvoiceDate)
cancelaciones_por_fecha = ordenes_canceladas['InvoiceDate'].value_counts()

fecha_con_mas_cancelaciones = cancelaciones_por_fecha.idxmax()
cantidad_cancelaciones_fecha = cancelaciones_por_fecha.max()

print(f"La fecha con más cancelaciones es {fecha_con_mas_cancelaciones} con {cantidad_cancelaciones_fecha} cancelaciones.")
```

```python
#Los 5 productos más caros
# Calcular el costo total de cada producto
minorista_limpieza_df['TotalCost'] = abs(minorista_limpieza_df['Quantity']) * minorista_limpieza_df['UnitPrice']

# Filtrar solo las filas con TotalCost positivo
productos_positivos = minorista_limpieza_df[minorista_limpieza_df['TotalCost'] > 0]

# Obtener los 5 productos más caros
top_5_mas_caros = productos_positivos[['TotalCost', 'Description']].sort_values(by='TotalCost', ascending=False).head(5)

print("Los 5 productos más caros del dataset son:")
print(top_5_mas_caros)
```

```python
# Los productos más caros, no tienen relación con los productos más cancelados.
```

```python
#Producto más vendido por país
df_filtered = minorista_limpieza_df.dropna()
df_filtered = df_filtered[(df_filtered['Quantity'] > 0) & (df_filtered['UnitPrice'] > 0)]

# Ahora puedes ejecutar el análisis de los productos más vendidos por país
top_selling_products_by_country = df_filtered.groupby(['Country', 'Description'])['Quantity'].sum()

# Ordenar dentro de cada país para obtener el producto más vendido en cada uno
top_selling_products_by_country = top_selling_products_by_country.groupby(level=0, group_keys=False).nlargest(1)

print(top_selling_products_by_country)
```

```python
#Màximo valor de transacción por pais
#Dar clic en el pais para validar el máximo valor de transacción
import plotly.express as px

# Filtrar los datos para eliminar valores negativos (posibles devoluciones) y agregar columna de valor total de cada transacción
df_filtered = minorista_limpieza_df[(minorista_limpieza_df['Quantity'] > 0) & (minorista_limpieza_df['UnitPrice'] > 0)].copy()
df_filtered['TotalValue'] = df_filtered['Quantity'] * df_filtered['UnitPrice']

#Obtener el máximo valor de transacción por país
max_transaction_per_country = df_filtered.groupby('Country')['TotalValue'].max().reset_index()

#Crear mapa
fig = px.choropleth(
    max_transaction_per_country,
    locations="Country",  # Columna con los países
    locationmode='country names',  # Modo de ubicación por nombre de país
    color="TotalValue",  # Columna con el valor para colorear los países
    color_continuous_scale="Viridis",  # Escala de color
    labels={'TotalValue': 'Máximo valor de transacción'},  # Etiquetas para la leyenda
    title="Mapa de calor: Máximo valor de transacción por país"
)
#Mostrar el gráfico
fig.update_geos(showcoastlines=True, coastlinecolor="Black")
fig.show()

```

```python
#Cantidad de clientes únicos por país
#Eliminar valores nulos en 'CustomerID' para evitar contar registros sin cliente
unique_customers_by_country = df_filtered.dropna(subset=['CustomerID']).groupby('Country')['CustomerID'].nunique()

print(unique_customers_by_country)
```

```python
#Nùmero de productos por transacción
temp_df = minorista_limpieza_df.groupby(['InvoiceNo', 'CustomerID'], as_index=False)['InvoiceDate'].count()
transaction_df = temp_df.rename(columns={'InvoiceDate': 'Number of products'})
transaction_df.head(10)
```
